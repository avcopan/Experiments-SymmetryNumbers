```python
from stereomolgraph import StereoCondensedReactionGraph
from stereomolgraph.algorithms.symmetry import (
    ext_sym_num,
    topological_symmetry_number,
)
from stereomolgraph.ipython import View2D
from stereomolgraph.stereodescriptors import (
    PlanarBond,
    Tetrahedral,
    TrigonalBipyramidal,
)

View2D.generate_bond_orders = False
View2D.color_planar_bond_changes = True
```

```python
methane_h_abs = StereoCondensedReactionGraph()

methane_h_abs.add_atom(0, "C")
methane_h_abs.add_atom(1, "H")
methane_h_abs.add_atom(2, "H")
methane_h_abs.add_atom(3, "H")
methane_h_abs.add_atom(4, "H")

methane_h_abs.add_bond(0, 1)
methane_h_abs.add_bond(0, 2)
methane_h_abs.add_bond(0, 3)
methane_h_abs.add_broken_bond(0, 4)
methane_h_abs.set_atom_stereo_change(broken=Tetrahedral((0, 1, 2, 3, 4), 1))

methane_h_abs.add_atom(5, "H")
methane_h_abs.add_formed_bond(5, 4)

methane_h_abs
```

```python
ethane_h_abs = StereoCondensedReactionGraph()

ethane_h_abs.add_atom(0, "C")
ethane_h_abs.add_atom(1, "H")
ethane_h_abs.add_atom(2, "H")
ethane_h_abs.add_atom(3, "H")
ethane_h_abs.add_atom(4, "C")
ethane_h_abs.add_bond(0, 1)
ethane_h_abs.add_bond(0, 2)
ethane_h_abs.add_bond(0, 3)
ethane_h_abs.add_bond(0, 4)
ethane_h_abs.set_atom_stereo(Tetrahedral((0, 1, 2, 3, 4), 1))

ethane_h_abs.add_atom(5, "H")
ethane_h_abs.add_bond(4, 5)
ethane_h_abs.add_atom(6, "H")
ethane_h_abs.add_bond(4, 6)
ethane_h_abs.add_atom(7, "H")
ethane_h_abs.add_broken_bond(4, 7)

ethane_h_abs.set_atom_stereo_change(broken=Tetrahedral((4, 0, 5, 6, 7), 1))

ethane_h_abs.add_atom(8, "H")
ethane_h_abs.add_formed_bond(7, 8)

ethane_h_abs
```

```python
ethene_h_abs = StereoCondensedReactionGraph()

ethene_h_abs.add_atom(0, "C")
ethene_h_abs.add_atom(1, "H")
ethene_h_abs.add_atom(2, "H")
ethene_h_abs.add_atom(4, "C")
ethene_h_abs.add_bond(0, 1)
ethene_h_abs.add_bond(0, 2)

ethene_h_abs.add_bond(0, 4)
ethene_h_abs.set_atom_stereo_change(formed=Tetrahedral((4, 0, 5, 6, 7), 1))

ethene_h_abs.add_atom(5, "H")
ethene_h_abs.add_bond(4, 5)
ethene_h_abs.add_atom(6, "H")
ethene_h_abs.add_bond(4, 6)
ethene_h_abs.add_atom(7, "H")
ethene_h_abs.add_formed_bond(4, 7)

ethene_h_abs.set_bond_stereo_change(broken=PlanarBond((1, 2, 0, 4, 5, 6), 0))

ethene_h_abs
```

```python
from pathlib import Path

import pandas as pd
from _pandas_patcher import PrintAsImageString
from IPython.display import HTML, display

cases = [
    ("methane_h_abs", methane_h_abs),
    ("ethane_h_abs", ethane_h_abs),
    ("ethene_h_abs", ethene_h_abs),
    ("tbp1", tbp1),
    ("tbp2", tbp2),
    ("tbp3", tbp3),
    ("tbp4", tbp4),
    ("tbp5", tbp5),
    ("tbp6", tbp6),
]

rows = [
    {
        "scrg": scrg,
        "top_sym_num": topological_symmetry_number(ts_graph),
        "ext_sym_num": ext_sym_num(ts_graph),
    }
    for _, scrg in cases
    for ts_graph in (scrg.ts(),)
]

df_ts_symmetry = pd.DataFrame(rows)
if not df_ts_symmetry.empty:
    df_ts_symmetry["top_sym_num"] = df_ts_symmetry["top_sym_num"].astype(
        "Int64"
    )
    df_ts_symmetry["ext_sym_num"] = df_ts_symmetry["ext_sym_num"].astype(
        "Int64"
    )

styled_ts_symmetry = (
    df_ts_symmetry[["scrg", "top_sym_num", "ext_sym_num"]]
    .style.hide(axis="index")
    .format(
        {
            "scrg": PrintAsImageString,
        },
        escape=None,
    )
    .set_properties(**{"text-align": "left", "vertical-align": "middle"})
    .set_table_styles(
        [
            {"selector": "table", "props": [("border-collapse", "collapse")]},
            {
                "selector": "th:not(:last-child), td:not(:last-child)",
                "props": [("border-right", "1px solid #b0b0b0")],
            },
            {"selector": "th", "props": [("text-align", "left")]},
        ]
    )
)

table_html = styled_ts_symmetry.to_html()
html_document = (
    "<!DOCTYPE html><html lang='en'><head><meta charset='utf-8'>"
    "<title>TS Symmetry Report</title></head><body>"
    f"<div style='overflow-x:auto'>{table_html}</div></body></html>"
)
project_root = next(
    candidate
    for candidate in (Path.cwd().resolve(), *Path.cwd().resolve().parents)
    if (candidate / "pyproject.toml").exists()
)
output_html_path = project_root / "reports" / "ts_sym_num_report.html"
output_html_path.parent.mkdir(parents=True, exist_ok=True)
output_html_path.write_text(html_document, encoding="utf-8")

display(HTML(f"<div style='overflow-x:auto'>{table_html}</div>"))
print(f"Saved HTML report to: {output_html_path.resolve()}")
```
