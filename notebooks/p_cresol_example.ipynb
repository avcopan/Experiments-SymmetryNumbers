```python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole

# Enable indices for all subsequent inline plots
IPythonConsole.drawOptions.addAtomIndices = True

smiles = "Cc1ccc(O)cc1"
rdmol = Chem.MolFromSmiles(smiles, sanitize=False)
rdmol.UpdatePropertyCache(strict=True)
rdmol = Chem.AddHs(rdmol)
Chem.SanitizeMol(rdmol)

rdmol
```

```python
from IPython.display import display
from stereomolgraph import StereoMolGraph
from stereomolgraph.algorithms import symmetry

graph = StereoMolGraph.from_rdmol(
    rdmol, stereo_complete=True, resonance=True, lone_pair_stereo=True
)
display(graph)
display(rdmol)

top_sym_num = symmetry.topological_symmetry_number(graph)
ext_sym_num = symmetry.external_symmetry_number(graph)
bnd_sym_num1 = symmetry.bond_symmetry_number(graph, bond=(0, 1))
bnd_sym_num2 = symmetry.bond_symmetry_number(graph, bond=(4, 5))
print(f"{top_sym_num = }")
print(f"{ext_sym_num = }")
print(f"{bnd_sym_num1 = }")
print(f"{bnd_sym_num2 = }")
```
